```python
import gymnasium as gym
import time

# pip installs:
# 1. pip install gymnasium
# 2. pip install swig
# 3. pip install "gymnasium[box2d]"

# Environments of my interest
ENVS = ["CarRacing-v3", "LunarLander-v3", "MountainCar-v0"] # Environment version v2 for `CarRacing` is deprecated. Using `CarRacing-v3` instead. Same v3 for LunarLander. MountainCar did not need changes

if __name__ == "__main__":
    # For CarRacing, you must specify render_mode="human" at creation
    for ENV_NAME in ENVS:
        if ENV_NAME == "CarRacing-v3":
            env = gym.make(ENV_NAME, render_mode="human")
        else:
            env = gym.make(ENV_NAME, render_mode="human")

        # Resetting the environment to get the first observation
        observation, info = env.reset()

        # Loop for 5000 steps (initial loop 1000 steps not satisfactory, choosing 5000)
        for _ in range(5000):
            # Render the environment
            env.render()

            # Random action from the environment's action space
            action = env.action_space.sample()

            # Perform the action and get the new state, reward, etc.
            observation, reward, terminated, truncated, info = env.step(action)

            # If the episode is over (e.g., car crashed, landed, or goal reached), reset it
            if terminated or truncated:
                print("Episode finished. Resetting.")
                observation, info = env.reset()
                
            time.sleep(0.01)

        # Clean up and close the environment
        env.close()
        print(f"Finished running {ENV_NAME}.")
```
